```
import numpy as np #import numpy lib
import random #import random lib
import math #import math lib
from matplotlib import pyplot as plt #import matplotlib


#define wt, ht array according to the conditions
def getRandomWeight():
    return np.random.uniform(low=40.0, high=90.0, size=100)

def getRandomHeight():
    return np.random.randint(140, 200, size=100)

#calculate bmi
def calculateBmi(weight, height):
    height = height / 100 #centimeters to meters
    return weight / height**2

#print bmi array
def printBmi(bmi):
    print("Bmi array")
    print(bmi)

    #print the first 10 elements of the bmi array
    print("First 10 elements of the bmi array.")
    print(bmi[0:10])

#bar chart
def barChart(bmi):
    # Count according to the weight
    underWeight = bmi[np.where(bmi <= 18.5)].size
    healthy = bmi[np.where((bmi >= 18.5) & (bmi < 25.0))].size
    overWeight = bmi[np.where((bmi >= 25.0) & (bmi < 30.0))].size
    obese = bmi[np.where((bmi >= 30.0))].size

    cases = ['underWeight', 'healthy', 'overWeight', 'obese']
    nums = [underWeight, healthy, overWeight, obese]

    plt.bar(cases, nums)
    plt.show()

#histogram
def histogram(bmi):

    plt.hist(bmi, bins=[0, 18.5, 25, 30, 50])
    plt.title("BMI")
    plt.xticks([0, 18.5, 25, 30, 50])
    plt.xlabel("BMI level")
    plt.ylabel("number of students")
    plt.show()

#pie chart
def pieChart(bmi):
    # Count according to the weight
    underWeight = bmi[np.where(bmi <= 18.5)].size
    healthy = bmi[np.where((bmi >= 18.5) & (bmi < 25.0))].size
    overWeight = bmi[np.where((bmi >= 25.0) & (bmi < 30.0))].size
    obese = bmi[np.where((bmi >= 30.0))].size

    cases = ['underWeight', 'healthy', 'overWeight', 'obese']
    nums = [underWeight, healthy, overWeight, obese]

    plt.pie(nums, labels=cases, autopct='%1.2f%%')
    plt.show()

#scatter plot
def scatterPlot(ht, wt):

    plt.scatter(ht, wt, color='b')
    plt.xlabel("Height Range")
    plt.ylabel("Weight Range")
    plt.title("BMI")
    plt.show()

#main function
def main():
    #call functions
    ht = getRandomHeight()
    wt = getRandomWeight()
    bmi = calculateBmi(wt, ht)
    printBmi(bmi)
    barChart(bmi)
    histogram(bmi)
    pieChart(bmi)
    scatterPlot(ht, wt)



#execute main function
if __name__=='__main__':
    main()
```

```
import numpy as np
# input data
A = [30,200,10,4]
B = [40,300,20,4]
C = [50,800,20,1]
D = [60,600,20,2]
E = [40,300,20,5]

data = np.array([A,B,C,D,E])
# population covariance matrix (N)
covMatrix = np.cov(data,bias=True)
print (covMatrix)


import seaborn as sn
import matplotlib.pyplot as plt
sn.heatmap(covMatrix, annot=True, fmt='g')
plt.show()

# sample covariance matrix (N-1)
covMatrix = np.cov(data,bias=False)
print (covMatrix)



import pandas as pd
data = {'A' : [30,200,10,4],
'B' : [40,300,20,4],
'C' : [50,800,20,1],
'D' : [60,600,20,2],
'E' : [40,300,20,5]
}
df = pd.DataFrame(data,columns=['A','B','C','D','E'])
# sample covariance matrix
covMatrix = pd.DataFrame.cov(df)
print (covMatrix)

import seaborn as sn
import matplotlib.pyplot as plt
sn.heatmap(covMatrix, annot=True, fmt='g')
plt.show()
```
